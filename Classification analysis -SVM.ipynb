```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score, classification_report

# Load the handwritten digits dataset
# This is a built-in dataset in scikit-learn
digits = datasets.load_digits()

# The data is a 2D array of pixel values for each image
# We need to reshape it into a 1D array (flattening the image)
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))

# Split the data into training and testing sets
# We use a 70/30 split, which is a common practice
X_train, X_test, y_train, y_test = train_test_split(
    data, digits.target, test_size=0.3, random_state=42
)

# Create an SVM classifier
# We use a kernel because the data isn't linearly separable
# The 'gamma' and 'C' parameters are important for tuning
# 'gamma' defines the influence of a single training example
# 'C' is a regularization parameter
clf = svm.SVC(gamma=0.001, C=100)

# Train the model on the training data
print("Training the SVM classifier...")
clf.fit(X_train, y_train)

# Make predictions on the test data
print("Making predictions...")
y_pred = clf.predict(X_test)

# Evaluate the model's performance
print("\nModel Evaluation:")
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Optional: Visualize a few predictions
_, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for ax, image, prediction, actual in zip(axes, X_test, y_pred, y_test):
    ax.set_axis_off()
    image = image.reshape(8, 8)
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
    ax.set_title(f"Pred: {prediction}\nActual: {actual}")
plt.show()
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score, classification_report

# Load the handwritten digits dataset
# This is a built-in dataset in scikit-learn
digits = datasets.load_digits()

# The data is a 2D array of pixel values for each image
# We need to reshape it into a 1D array (flattening the image)
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))

# Split the data into training and testing sets
# We use a 70/30 split, which is a common practice
X_train, X_test, y_train, y_test = train_test_split(
    data, digits.target, test_size=0.3, random_state=42
)

# Create an SVM classifier
# We use a kernel because the data isn't linearly separable
# The 'gamma' and 'C' parameters are important for tuning
# 'gamma' defines the influence of a single training example
# 'C' is a regularization parameter
clf = svm.SVC(gamma=0.001, C=100)

# Train the model on the training data
print("Training the SVM classifier...")
clf.fit(X_train, y_train)

# Make predictions on the test data
print("Making predictions...")
y_pred = clf.predict(X_test)

# Evaluate the model's performance
print("\nModel Evaluation:")
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Optional: Visualize a few predictions
_, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for ax, image, prediction, actual in zip(axes, X_test, y_pred, y_test):
    ax.set_axis_off()
    image = image.reshape(8, 8)
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
    ax.set_title(f"Pred: {prediction}\nActual: {actual}")
plt.show()
```
